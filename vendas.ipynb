```python
import pandas as pd

ds = pd.read_csv('vendas.csv', delimiter = ',')
```

```python
#Criando a coluna "Faturamento Total"

ds['Faturamento Total'] = ds['Preco_Unitario'] * ds['Quantidade']

ds['Data'] = pd.to_datetime(ds['Data']) #Transformando a coluda data num valor cronol√≥gico
ds.info()
```

```python
ds.to_csv('vendas_tratadas.csv', index= False)
```
